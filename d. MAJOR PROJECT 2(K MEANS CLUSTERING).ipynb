```python
#main task- find out no. of cluster

import numpy as np
np.sqrt(178)
```

```python
#elbow method
from sklearn.cluster import KMeans
k=range(2,15)  #range between 2 and 14
sse=[] #blank list
for i in k:
  model_demo= KMeans(n_clusters= i, random_state=0)
  model_demo.fit(x)
  sse.append(model_demo.inertia_)  #inertia calc the squared error
plt.scatter(k,sse)
plt.plot(k,sse)
```

```python
#2. SILHOUETTE score method
from sklearn.metrics import silhouette_score
k=range(2,15)
for i in k:
  model_demo= KMeans(n_clusters= i,random_state=0)
  model_demo.fit(x)
  y_pred= model_demo.predict(x)
  print(f"{i} clusters, Score= {silhouette_score(x,y_pred)}")
  plt.bar(i,silhouette_score(x,y_pred))
```

```python
#7. applying cluster
k=5
from sklearn.cluster import KMeans
model= KMeans(n_clusters=k,random_state=0)
model.fit(x)
```

```python
y= model.predict(x)    #predicted op
print(y)
```

```python
y.size
```

```python
print(x[y== 1,1])
#so the first '1' is cluster no 1 and the second '1' is column index 1
#the value of input,when cluster 1 is selected and column index 1 selected
```

```python
#FINAL VISUALISATION
plt.figure(figsize = (10,5))
for i in range(k):
  plt.scatter(x[y == i,0],x[y == i,1],label = f'Cluster {i}')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s = 300,c = 'black',
            label = 'Centroids')
plt.legend()
```
